## Converting DataFrame columns to appropriate data types

```python
cafe['Quantity'] = pd.to_numeric(cafe['Quantity'], errors='coerce')
cafe['Price Per Unit'] = pd.to_numeric(cafe['Price Per Unit'], errors='coerce')
cafe['Total Spent'] = pd.to_numeric(cafe['Total Spent'], errors='coerce')
cafe['Transaction Date']=pd.to_datetime(cafe['Transaction Date'], errors='coerce')
```

```python
cafe.dtypes
```

## Replace ERROR UNKNOWN values to NaN

```python
cafe.replace(['ERROR','UNKNOWN'],np.nan,inplace=True)
```

## Filling missing price values based on item name


```python
invalid_items=['UNKNOWN',np.nan, 'ERROR']
valid_rows = cafe[~cafe['Item'].isna()]
price_map = valid_rows.drop_duplicates(subset=['Item'])
price_map = price_map.set_index('Item')['Price Per Unit'].to_dict()
price_map
```

```python
cafe[cafe['Price Per Unit'].isna()]
```

```python
cafe['Price Per Unit'] = cafe['Price Per Unit'].fillna(cafe['Item'].map(price_map))
```

```python
cafe
```

## Fill price per unit with total spent/quantity

```python
cafe['Price Per Unit'] = cafe['Price Per Unit'].fillna(cafe['Total Spent']/cafe['Quantity'])
```

```python
cafe[cafe['Price Per Unit'].isna()]
```

```python
cafe.info()
```

## Fill price per unit with mode

```python
cafe['Price Per Unit']=cafe['Price Per Unit'].fillna(cafe['Price Per Unit'].mode().iloc[0])
```

```python
cafe.info()
```

## Fill null values in Quantity column

```python
cafe[cafe['Quantity'].isna()]
```

#### fill using total spent/price

```python
cafe['Quantity']=cafe['Quantity'].fillna(cafe['Total Spent']/cafe['Price Per Unit'])
```

```python
cafe[cafe['Quantity'].isnull()]
```

```python
cafe.loc[cafe['Quantity'].isnull(),'Quantity']=cafe.groupby('Item')['Quantity'].transform("mean")
```

## Fill missing values in 'Item' based on price

```python
type(cafe)
```

```python
### Using " Boolean Index "
```

```python
def item_basedon_price(df):
    df.loc[(df['Price Per Unit']==2) & (df.Item.isna()),'Item']='Coffee'
    df.loc[(df['Price Per Unit']==3) & (df.Item.isna()),'Item']='Cake'
    df.loc[(df['Price Per Unit']==1) & (df.Item.isna()),'Item']='Cookie'
    df.loc[(df['Price Per Unit']==5) & (df.Item.isna()),'Item']='Salad'
    df.loc[(df['Price Per Unit']==4) & (df.Item.isna()),'Item']='Smoothie'
    df.loc[(df['Price Per Unit']==3) & (df.Item.isna()),'Item']='Juice'
    df.loc[(df['Price Per Unit']==1.5) & (df.Item.isna()),'Item']='Tea'
    return df
```

```python
cafe = item_basedon_price(cafe)
```

## Fill Missing values in 'Location' using probability random distribution

```python
cafe[cafe['Location'].isna()]
```

```python
probs_loc = cafe['Location'].value_counts(normalize=True)
probs_loc
```

```python
probs_loc.index
```

```python
cafe.loc[cafe['Location'].isna(),'Location']=np.random.choice(probs_loc.index, size=cafe.Location.isna().sum(),p=probs_loc.values)
```

## Detecting Outliers

```python
sns.set(style="whitegrid")  # Let's set the graphic style

# Let's create boxplots for different variables
fig, axes = plt.subplots(2, 2, figsize=(20, 18))

sns.boxplot(x=cafe["Quantity"], ax=axes[0, 0])  # Boxplot for quantity
axes[0, 0].set_title("Boxplot for Quantity")

sns.boxplot(x=cafe["Price Per Unit"], ax=axes[0, 1])  # Boxplot for Price Per Unit
axes[0, 1].set_title("Boxplot for Price Per Unit")

sns.boxplot(x=cafe["Total Spent"], ax=axes[1, 0])  # Boxplot for Total Spent
axes[1, 0].set_title("Boxplot for Total Spent")

plt.show()
```

```python
# Set the figure style
sns.set(style="darkgrid")

# Creating figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 15))

# Plot 1: Quantity vs Price Per Unit
sns.scatterplot(x=cafe['Quantity'], y=cafe['Price Per Unit'], alpha=0.5, ax=axes[0, 0])
axes[0, 0].set_title("Quantity vs Price Per Unit")

# Plot 2: Location vs Total Spent
sns.boxplot(x=cafe['Location'], y=cafe['Total Spent'], ax=axes[0, 1])
axes[0, 1].set_title("Location vs Total Spent")
axes[0, 1].tick_params(axis='x', rotation=45)

# Plot 3: Payment Method vs Total Spent
sns.boxplot(x=cafe['Payment Method'], y=cafe['Total Spent'], ax=axes[1, 0])
axes[1, 0].set_title("Payment Method vs Total Spent")
axes[1, 0].tick_params(axis='x', rotation=45)

# Plot 4: Heatmap of Correlation Matrix
corr = cafe.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[1, 1])
axes[1, 1].set_title("Correlation Heatmap")

# Adjusting layout and showing plots
plt.tight_layout()
plt.show()
```
