## API Runner

```python
def api_runner():
    global df
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1',
      'limit':'15',
      'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': '148461e7-0ff9-493f-80c7-075bbe93b365',
    }
    
    session = Session()
    session.headers.update(headers)
    
    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      #print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)

    df = pd.json_normalize(data['data'])
    df['time stamp'] = pd.to_datetime('now')
    if not os.path.isfile(r'C:\Users\shrav\latest_python\WEB API\API.csv'):
        df.to_csv(r'C:\Users\shrav\latest_python\WEB API\API.csv', header='column_names')
    else:
        df.to_csv(r'C:\Users\shrav\latest_python\WEB API\API.csv', mode='a', header=False)
```

## Script to run the API runner automatically

```python
import os
from time import time
from time import sleep

# 333 is the max number of requests allowed in the given site
for i in range(2):
    api_runner()
    print('API runner completed successfully')
    # run for every 60 secs
    sleep(5)
    
```
