### Putting it all together

```python
results = []
for r in result_items:
    items = {}
    items['Address'] = r.get('Property', {}).get('Address', {}).get('AddressText', 'N/A')
    items['Bed rooms'] = r.get('Building', {}).get('Bedrooms', 'N/A')
    items['Bathrooms'] = r.get('Building', {}).get('BathroomTotal', 'N/A')
    items['Agent Name'] = r.get('Individual', [{}])[0].get('Name', 'N/A')
    items['Area Code'] = r.get('Individual', [{}])[0].get('Phones', [{}])[0].get('AreaCode', 'N/A')
    items['Phone Number'] = r.get('Individual', [{}])[0].get('Phones', [{}])[0].get('PhoneNumber', 'N/A')
    items['Price'] = r.get('Property', {}).get('Price', 'N/A')
    results.append(items)
```
