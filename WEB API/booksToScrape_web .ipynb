```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python

proceed = True  

# Starting page number
current_page = 1  

data = []  

# Loop through pages until a 404 error is encountered
while proceed:
    url = 'https://books.toscrape.com/catalogue/page-' + str(current_page) + '.html'
    
    # Send an HTTP GET request to the URL
    page = requests.get(url)
    
    # Ensure correct text encoding (UTF-8)
    page.encoding = "utf-8"
    
    # Parse the webpage content with BeautifulSoup
    soup = BeautifulSoup(page.text, 'html.parser')
    
    if soup.text == '\n404 Not Found\n\n404 Not Found\nnginx/1.21.6\n\n\n':
        proceed = False  # Stop the loop if the page is not found
    else:
        # Find all book elements on the current page
        all_books = soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
        
        # Loop through each book and extract details
        for book in all_books:
            item = {} 
            
            # Extract book title (from the 'alt' attribute of the <img> tag)
            item['title'] = book.find('img').attrs['alt']
            
            # Extract book link (concatenating the base URL with the relative link)
            item['link'] = 'https://books.toscrape.com/catalogue/' + book.find('a').attrs['href']
            
            # Extract book price
            item['price'] = book.find('p', class_='price_color').text
            
            # Extract stock availability (strip to remove extra spaces)
            item['stock'] = book.find('p', class_='instock availability').text.strip()
            
            data.append(item)
        current_page += 1  

        
    
    
```
