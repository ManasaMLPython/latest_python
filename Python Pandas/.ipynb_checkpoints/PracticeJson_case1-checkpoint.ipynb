## Flatten nested json file into tabular format using json_normalize

```python
pd.json_normalize(data=data, sep="_")
```

## normalizing a specific column of the json

```python
pd.json_normalize(data=data, record_path="genres", meta=["title"]).head(10)
```

```python
pd.json_normalize(data=data, record_path="genres", meta=["title","id"], record_prefix="genre_").head(10)
```
