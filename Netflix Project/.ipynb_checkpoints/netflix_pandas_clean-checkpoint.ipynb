```python
import pandas as pd
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Fill nulls with unknown in 'director','cast' and 'country'columns
df[['director','cast','country']] = df[['director','cast','country']].fillna('unknown')
```

```python
df.isna().sum()
```

```python
# fill rating nulls with real world data based on internet
df[df['rating'].isna()]
```

```python
df.loc[5989,'rating'] = 'TV‑PG'
df.loc[6827,'rating'] = 'TV‑14'
df.loc[7312,'rating'] = 'TV‑PG'
df.loc[7537,'rating'] = 'PG-13'
```

```python
# duration
df[df['duration'].isna()]
```

```python
#observed missplaced value in differnt column so correcting it
df.loc[5541,'duration']=df.loc[5541,'rating']
df.loc[5794,'duration']=df.loc[5794,'rating']
df.loc[5813,'duration']=df.loc[5813,'rating']
```

```python
#correcting the rating to real value
df.loc[5541,'rating']='TV-MA'
df.loc[5794,'rating']='TV-MA'
df.loc[5813,'rating']='TV-MA'
```

```python
df.dtypes
```

```python
# changing date_added to date time
df['date_added'].unique()
```

```python
df['date_added']=pd.to_datetime(df['date_added'], format='mixed')
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.head(10)
```

```python
# change duration to int by extracting the numeric part only
df['duration'] = df['duration'].str.replace(r'[^0-9]','',regex=True)
```

```python
df['duration'] = pd.to_numeric(df['duration'])
```

```python
df.to_csv('netflix_cleaned.csv')
```
