# Indexing and Slicing with reindexing()

```python
s=pd.read_csv("summer.csv")
```

```python
s.loc[[0,5,30000],["Athlete","Medal"]]
```

```python
s.loc[[0,5,30000,40000],["Athlete","Medal"]]
```

```python
s.reindex(index=[0,5,30000,40000],columns=["Athlete","Medal","Age"])
```

# Advanced Indexing and Slicing

### case 1: getting the first 5 rows and rows 354 and 765

```python
rows=list(range(5))+[354,765]
rows
```

```python
summer.iloc[rows]
```

### case2: Getting the first 3 columns and the columns "Gender" and "Event"

```python
cols=list(summer.columns[:3])+["Gender", "Event"]
cols
```

```python
summer.loc[:,cols]
```

### case3: Combining position and label based indexing: Rows at position 200, 300 and columns "Athlete" and "Medal" 

```python
summer1=pd.read_csv("summer.csv")
summer1.head()
```

```python
summer1.iloc[[200,300]].loc[:,["Athlete","Medal"]]
```

```python
summer1.loc[[200,300],["Athlete","Medal"]]
```

### case3: Combining position and label based indexing: Rows "PHELPS Michael" and position arguments 4 and 6

```python
cols=summer.columns[[4,6]]
cols
```

```python
summer.loc["PHELPS, Michael",cols]
```
