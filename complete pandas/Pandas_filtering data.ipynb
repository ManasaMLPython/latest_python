```python
# name starts with A ends with d
bios[(bios['name'].str.startswith('A')) & (bios['name'].str.endswith('d'))]['name']
```

```python
# find athlets that born in cities starts with a vowel
bios[bios['born_city'].str.contains(r'^[AEIOUaeiou]',na=False)][['name','born_city']]
```

```python
# find athlets with names that contain exactly 2 vowels
import re
def vowel_count(name):
    return len(re.findall(r'[AEIOUaeiou]',name))

bios[bios['name'].apply(vowel_count)==2]['name']
```

```python
# find athlets name ending in son/sen
bios[bios['name'].str.contains(r'son$|sen$')]['name']
```

```python
# find athlet born in a year starting with '19'
bios[bios['born_date'].str.contains(r'^19',na=False)][['name','born_date']]
```

```python
# find athlets with names that do not contain any vowels
bios[bios['name'].apply(vowel_count)==0]['name']
```

```python
# find athlets with names contain a hyphen or an apostrophe
bios[bios['born_date'].str.contains(r"[-']",na=False)]['name']
```
