# Aggregating

```python
coffee = pd.read_csv('./warm_up_data/coffee.csv')
```

```python
coffee
```

```python
import numpy as np
coffee['price'] = np.where(coffee['Coffee Type']=='Espresso',3.99,5.99)
```

```python
coffee['Revenue']=coffee['Units Sold']*coffee['price']
```

```python
coffee.loc[2:4, 'Units Sold']=np.nan
```

```python
coffee
```

```python
coffee['Units Sold']=coffee['Units Sold'].interpolate()
```

```python
coffee
```

```python
coffee.groupby(['Coffee Type', 'Day'])['Units Sold'].sum()
```

```python
coffee.groupby('Coffee Type')['Units Sold'].mean()
```

```python
coffee
```

```python
pivot = coffee.pivot(index='Day', columns='Coffee Type', values='Revenue')
pivot
```

```python
pivot.sum()
```

```python
pivot.sum(axis=1)
```
