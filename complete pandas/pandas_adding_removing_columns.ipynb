```python
bios = pd.read_csv('./data/bios.csv')
```

```python
bios.head()
```

```python
bios_new= bios.copy()
bios_new['year']=bios_new['born_date'].str.split('-').str[0]
```

```python
bios_new
```

```python
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]
```

```python
bios_new
```

```python
bios_new.query('first_name == "Bill"')
```

```python
bios_new['born_date_time']=pd.to_datetime(bios_new['born_date'],errors='coerce', format='%Y-%m-%d')
```

```python
bios_new.info()
```

```python
bios_new['year']=bios_new['born_date_time'].dt.year.astype('Int64')
```

```python
bios_new
```

```python
bios_new.to_csv('./data/bios_new.csv')
```

```python
bios_new['height_category']=bios_new['height_cm'].apply(lambda x: 'short' if x<165 else 'average' if x<185 else 'tall')
```

```python
bios_new
```

```python
def categorize_athlet(row):
    if row['height_cm']<175 and row['weight_kg']<70:
        return 'Lightweight'
    elif row['height_cm']<185 and row['weight_kg']<80:
        return 'middleweight'
    else:
        return 'heavyweight'
```

```python
bios_new['category']=bios.apply(categorize_athlet, axis=1)
```
