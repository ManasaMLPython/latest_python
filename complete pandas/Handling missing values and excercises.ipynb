# Excercises
### 1. Fill missing "Units Sold" with the average units sold for that coffee type.

```python
coffee['Units Sold']=coffee.groupby('Coffee Type')['Units Sold'].transform(lambda x: x.fillna(x.mean()))
```

```python
coffee
```

### 2. Fill missing "Revenue" using Units Sold × price:

```python
coffee['Revenue']=coffee['Revenue'].fillna(coffee['Units Sold']*coffee['price'])
```

```python
coffee
```

### 3. Drop rows with any missing values:

```python
coffee.dropna(inplace=True)
```

```python
coffee
```

# Data Exploration
### 5. Total units sold by coffee type:

```python
coffee.groupby('Coffee Type')['Units Sold'].sum()
```

### 6. Total revenue per day:

```python
coffee.groupby('Day')['Revenue'].sum().sort_values()
```

### 7. Average price of all coffee types:

```python
coffee.groupby('Coffee Type')['price'].mean()
```

### 8. Day with the highest total revenue:

```python
coffee.groupby('Day')['Revenue'].sum().idxmax()
```

# Grouping and Aggrega
### 9. Group by day — total units and revenue:

```python
coffee.groupby('Day')[['Units Sold','Revenue']].sum()
```

### 10. Group by coffee type — average units and max revenue:

```python
coffee.groupby('Coffee Type').agg({'Units Sold':'mean', 'Revenue':'max'})
```

# Feature Engineering & Cleanup

```python
coffee
```

### 12. Check if "Revenue per Unit" ≠ price:

### 13. Add "Is Weekend" column:

### 14. Sort by Revenue descending:

### 16. Coffee type with highest average revenue per unit:

### 17. Days when Espresso sold more units than Latte:

### 17.  Days with total revenue > 300:

### 18. Revenue in top 10%:

```python
quantile = coffee['Revenue'].quantile(0.9)
quantile
```

```python
coffee[coffee['Revenue']>quantile]
```

### 19. Coffee type with increasing sales:

```python
coffee['sales diff']=coffee.groupby('Coffee Type')['Units Sold'].diff()
```

```python
coffee.groupby('Coffee Type')['sales diff'].apply(lambda x: x>0)
```

```python
coffee.groupby('Coffee Type')['sales diff'].apply(lambda x: (x>0).sum())
```
