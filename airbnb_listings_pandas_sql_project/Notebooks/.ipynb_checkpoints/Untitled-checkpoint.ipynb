# Handle Missing Values

### 1. Handle missing values in 'name' column

```python
abnb[abnb.name.isna()].head()
```

#### Backfill with room type or neighborhood context

```python
abnb['name'] = abnb['name'].fillna(abnb['room_type'] + ' in ' + abnb['neighbourhood'])
```

```python
abnb.loc[3703:]
```

```python
abnb.info()
```

## 2. Fill missing values in 'host_name'

```python
abnb['host_name'] = abnb['host_name'].fillna('unknown') 
```

```python
abnb.isna().sum()
```

## 3. Fill missing values in 'last_review'

##### Convert last_review column to datetime

```python
abnb['last_review'] = pd.to_datetime(abnb['last_review'], errors = 'coerce')
```
